```python
import os
import numpy as np
import shutil
```

```python
def splitImg(xml_dir,origin_img,img_dir):
    xmlFiles = os.listdir(xml_dir)
    for xml in xmlFiles: 
        originimgPath = origin_img + xml[:-4] + ".jpg"
        newimgPath = img_dir + xml[:-4] + ".jpg"
        print(shutil.copy(originimgPath,newimgPath))
```

```python
folder_list= ["train","val","test"]
#base_dir
base_dir = "./swim_voc/"
origin_img = base_dir + "JPEGImages/" 
for i in range(3):
    folderName = folder_list[i]
    xml_dir = base_dir + folderName + "/Annotations/"
    if not os.path.exists(xml_dir):
        os.makedirs(xml_dir)
    img_dir = base_dir + "/images/" + folderName + "/"
    if not os.path.exists(img_dir):
        os.makedirs(img_dir)
    print(img_dir)
    splitImg(xml_dir,origin_img,img_dir)
```
