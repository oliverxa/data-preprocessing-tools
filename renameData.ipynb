```python
import os, shutil
import random
import numpy as np
```

```python
def renameData(xmlDir, imgDir):
    xmlFiles = os.listdir(xmlDir)
    total = len(xmlFiles)
    cur = 0
    for xml in xmlFiles:
        cur += 1
        if cur % 500 == 1:
            print("Total/cur:", total, "/", cur)
        imgPath = imgDir + xml[:-4] + ".jpg"
        
        outName = ("%08d" % (cur))
        
        outXMLPath = ("%s/%s.xml" % (xmlDir,outName))
        outImgPath = ("%s/%s.jpg" % (imgDir,outName))
        
        os.rename(xmlDir+xml,outXMLPath)
        os.rename(imgPath,outImgPath)
        
    
    print("picker number:",cur)
```

```python
xmlDir = "./swim_voc/Annotations/"
imgDir = "./swim_voc/JPEGImages/"
```

```python
renameData(xmlDir, imgDir)
```

```python
import xml.etree.ElementTree as ET
```

```python
def change_xml(xml_path):
    filelist = os.listdir(xml_path)
    print(filelist)
    # 打开xml文档
    for xmlfile in filelist:
        doc = ET.parse(xml_path+xmlfile)
        root = doc.getroot()
        sub1 = root.find('filename')  #找到filename标签，
        sub1.text = xmlfile     #修改标签内容
        
        doc.write(xml_path+xmlfile)
```

```python
change_xml(xmlDir)
```
