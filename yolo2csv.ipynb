```python
import os
```

```python
yolo_path = "./swim12/"
abspath = os.path.abspath('swim12')
```

```python
data_t = []
for file in os.listdir(yolo_path):
    if file[-4:] == ".jpg":
        data_t.append(file[:-4])
        #print(file)
    #print(abspath+'/'+file)
```

```python
len(data_t),data_t[0]
```

```python
import cv2
img = cv2.imread(yolo_path+"0.jpg")
width, height, _ = img.shape
```

```python
labels = ['person']
```

```python
data_list = []
for num in data_t:
    obj = [abspath+"/"+num+".jpg"]
    with open(yolo_path + num + ".txt") as f:
        for line in f.readlines():
            x = float(line.split(" ")[1])*width
            y = float(line.split(" ")[2])*height
            w = float(line.split(" ")[3])*width
            h = float(line.split(" ")[4])*height
            #print(x,y,w,h)
            
            xmin = int(x - w/2)
            ymin = int(y - h/2)
            xmax = int(x + w/2)
            ymax = int(y + h/2)
            
            #print(xmin,ymin,xmax,ymax)
            obj += str(xmin),str(ymin),str(xmax),str(ymax),"person"
            print(obj)
        data_list.append(obj)
        #print(abspath+"/"+num+".jpg"+','+','.join(obj))
```

```python
len(data_list),data_list[0]
```

data csv

```python
import pandas as pd
swim_csv=pd.DataFrame(columns=None,index=None,data=data_list)
```

```python
swim_csv.to_csv("swim_data.csv",columns=None,index=None,header=None)
```
